### I want to see if the paircoutns are ordered or unordered

```python
import numpy as np
from Corrfunc.theory.DDsmu import DDsmu

N = 100
L = 1000.0
np.random.seed(42)  # for reproducibility
x = np.random.uniform(0, L, N)
y = np.random.uniform(0, L, N)
z = np.random.uniform(0, L, N)

# Bins covering all possible separations
max_sep = L * np.sqrt(3) / 2.0  # ~866.025
s_bins = np.array([0.0, max_sep + 1e-6])  # include all pairs
mu_max = 1.0
nmu_bins = 1

res = DDsmu(autocorr=1, nthreads=4, binfile=s_bins,
            mu_max=mu_max, nmu_bins=nmu_bins,
            X1=x, Y1=y, Z1=z,
            periodic=True, boxsize=L,
            verbose=False)

total_pairs = res['npairs'][0]
print(f"Total pairs from DDsmu: {total_pairs}")
print(f"N(N-1) = {N*(N-1)}")
print(f"N(N-1)/2 = {N*(N-1)/2}")
```
